```python
import pandas as pd
import numpy as np

# Create a DataFrame with RMSE values for each model
rmse_data = {
    "Plant": ["sak", "Tan", "pava", "phd", "rew", "bha", "dhu", "cha", "kam", "kur", "man"],
    "S1": [15.57, 25.52, 15.61, 15.45, 15.83, 17.39, 17.54, 11.03, 26.46, 14.96, 13.28],
    "S2": [14.21, 22.537, 11.72, 17.99, 23.72, 16.37, 16.29, 13.71, 25.14, 12.52, 14.74],
    "S3": [11.29, 24.81, 10.37, 15.27, 14.80, 15.43, 15.46, 9.81, 23.85, 10.92, 12.14],
    "Bi-LSTM": [17.82, 24.91, 14.47, 14.18, 18.60, 14.79, 14.76, 10.69, 23.14, 11.14, 15.32],
    "Persistence": [27.89, 31.92, 30.44, 27.76, 27.89, 27.92, 28.54, 27.44, 34.38, 28.75, 28.17]
}

rmse_df = pd.DataFrame(rmse_data)

# Function to calculate FSI for each row in the DataFrame
def calculate_fsi(row):
    proposed_model_rmse = min(row[["S1", "S2", "S3", "Bi-LSTM"]])
    persistence_rmse = row["Persistence"]
    fsi = 1 - (proposed_model_rmse / persistence_rmse)
    return fsi

# Calculate FSI for each plant and store the result in a new column
rmse_df["FSI"] = rmse_df.apply(calculate_fsi, axis=1)

# Display the resulting DataFrame with FSI values
print(rmse_df)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Find the best performed model for each plant
rmse_df['Best Model'] = rmse_df[['S1', 'S2', 'S3', 'Bi-LSTM']].idxmin(axis=1)
rmse_df['Best RMSE'] = rmse_df[['S1', 'S2', 'S3', 'Bi-LSTM']].min(axis=1)

# Create a bar plot for RMSE and FSI values
fig, ax1 = plt.subplots(figsize=(12, 6))

bar_width = 0.15
x = np.arange(len(rmse_df))

ax1.bar(x - bar_width * 2, rmse_df['S1'], width=bar_width, label='S1', alpha=0.8)
ax1.bar(x - bar_width, rmse_df['S2'], width=bar_width, label='S2', alpha=0.8)
ax1.bar(x, rmse_df['S3'], width=bar_width, label='S3', alpha=0.8)
ax1.bar(x + bar_width, rmse_df['Bi-LSTM'], width=bar_width, label='Bi-LSTM', alpha=0.8)
ax1.bar(x + bar_width * 2, rmse_df['Persistence'], width=bar_width, label='Persistence', alpha=0.8)

# Mark the best RMSE values
for i, row in rmse_df.iterrows():
    ax1.text(i - bar_width * 2, row['S1'] + 1, '*' if row['Best Model'] == 'S1' else '', ha='center', fontsize=12, color='blue')
    ax1.text(i - bar_width, row['S2'] + 1, '*' if row['Best Model'] == 'S2' else '', ha='center', fontsize=12, color='blue')
    ax1.text(i, row['S3'] + 1, '*' if row['Best Model'] == 'S3' else '', ha='center', fontsize=12, color='blue')
    ax1.text(i + bar_width, row['Bi-LSTM'] + 1, '*' if row['Best Model'] == 'Bi-LSTM' else '', ha='center', fontsize=12, color='blue')

ax1.set_xlabel('Plants')
ax1.set_ylabel('RMSE')
ax1.set_title('RMSE and FSI Values for Each Plant')
ax1.set_xticks(x)
ax1.set_xticklabels(rmse_df['Plant'])
ax1.legend()

# Add a second y-axis for FSI values
ax2 = ax1.twinx()
ax2.plot(x, rmse_df['FSI'], linestyle='-', marker='o', linewidth=2, label='FSI', color='red')
ax2.set_ylabel('FSI')
ax2.legend(loc='upper right')

fig.tight_layout()
plt.show()
```
